```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
top_movie = df.sort_values(by='Rating', ascending=False).head(10)
top_movie
```

```python
sns.histplot(data=top_movie, x="Year", hue="Rating", multiple="stack")
plt.title('Top 10 Movies Rating by Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Rating', fontsize=14)
plt.xticks(ticks=sorted(top_movie['Year'].unique()), labels=sorted(top_movie['Year'].unique()))
plt.tight_layout()
plt.show()
```

```python
genre_counts = df['Genre'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'count']

top_n_genres = genre_counts.head(6)
top_n_genres
```

```python
plt.figure(figsize=(4,4))
plt.pie(top_n_genres['count'],labels=top_n_genres['Genre'],autopct='%1.1f%%',startangle=140,colors=sns.color_palette('pastel'))
plt.title('Movie Genre Distribution', fontsize=20)
plt.axis('equal')
plt.show()
```

```python
director_avg_rating = df.groupby('Director')['Rating'].mean().reset_index()

director_avg_rating = director_avg_rating.sort_values(by='Rating', ascending=False)

top_directors = director_avg_rating.head(10)
top_directors
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=top_directors, x='Rating', y = 'Director', hue='Director', palette='viridis',)

plt.title('Top Directors by Average Rating', fontsize=30)
plt.xlabel('Average Rating', fontsize=14)
plt.ylabel('Director', fontsize=30)

plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Rating', y='Votes')

plt.title('Votes vs. Rating', fontsize=30)
plt.xlabel('Rating', fontsize=25)
plt.ylabel('Votes', fontsize=30)

plt.show()
```

```python
actor_counts = df['Actor 1'].value_counts().reset_index()
actor_counts.columns = ['Actor', 'MovieCount']

top_n_actors = actor_counts.head(10)
top_n_actors
```

```python
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')

df = df.dropna(subset=['Rating', 'Votes'])

filtered_df = df[(df['Rating'] > 8) & (df['Votes'] > 100)]

filtered_df.head(10)
```

```python
plt.figure(figsize=(15,6))
ax = sns.barplot(data=filtered_df, x='Name', y='Votes', hue='Rating', dodge=False, width=0.5, palette='muted')

ax.tick_params(axis='x', labelrotation=90)  
ax.legend(loc='upper right')
ax.set_xlabel('Movie Name',fontsize=16)
ax.set_ylabel('Votes',fontsize=16)
ax.set_title('Movies with rating greater than 8 and votes greater than 10000')

plt.show()
```

```python
df['Duration'] = df['Duration'].str.replace('min', '').str.strip().astype(int)

df['Year'] = df['Year'].astype(int)

plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Year', y='Duration', errorbar=None)
plt.xlabel('Year')
plt.ylabel('Duration in minutes')
plt.title('Duration of movies by year')
plt.xticks(np.arange(1917, 2023, 5))
plt.show()
```

```python
df['Genre'] = df['Genre'].str.split(',')

genre_df = df.explode('Genre')
genre_df
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=genre_df, x='Genre', order=genre_df['Genre'].value_counts().index, color='skyblue')

plt.title('Number of Movies for Each Genre', fontsize=16)
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Number of movies', fontsize=14)
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
average_rating_by_genre = genre_df.groupby('Genre')['Rating'].mean().reset_index()
average_rating_by_genre = average_rating_by_genre.sort_values(by='Rating', ascending=False)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=average_rating_by_genre, y="Rating", x="Genre", hue="Genre", palette="coolwarm", legend=False)

plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.title('Average rating of movies in each genre')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()

trans_data = df.drop(['Name'], axis=1)
trans_data['Director'] = labelencoder.fit_transform(df['Director'])
trans_data['Actor 1'] = labelencoder.fit_transform(df['Actor 1'])
trans_data['Actor 2'] = labelencoder.fit_transform(df['Actor 2'])
trans_data['Actor 3'] = labelencoder.fit_transform(df['Actor 3'])
trans_data['Genre'] = labelencoder.fit_transform(df['Genre'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x))
trans_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
sc_data = scaler.fit_transform(trans_data)
sc_df = pd.DataFrame(sc_data,columns=trans_data.columns)
sc_df.head()
```
